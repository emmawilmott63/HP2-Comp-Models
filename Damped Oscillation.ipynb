```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

### Motion of A Damped Oscillating System

Experimental Constants

```python
# mass of object (kg)
m = 0.029
# spring constant of spring in system (N/m)
k = 3
# initial position of object (m)
x = 0
# unstretched position of spring (m)
xrest = 0
# initial displacement of spring from unstretched position (m)
dx = 0
# air resistance constant
b = 0.01
```

Motion of a Vertical Damped Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.01
ar = b*v*abs(v)
a = (-k*dx - m*g - ar)/m

while t < (2):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    ar = b*v*abs(v)
    a = (-k*dx - m*g - ar)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

Motion of a Horizontal Damped Oscillator

```python
v = 0
t = 0
dt = 0.01
ar = b*v*abs(v)
a = (-k*dx - ar)/m

while t < (2):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    ar = b*v*abs(v)
    a = (-k*dx - ar)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

### Energy of a Damped Oscillating System

Experimental Constants

```python
# mass of object (kg)
m = 0.029
# spring constant of spring in system (N/m)
k = 3
# initial position of object (m)
x = 0
# unstretched position of spring (m)
xrest = 0
# initial displacement of spring from unstretched position (m)
dx = 0
# air resistance constant
b = 0.01
```

Energy of a Vertical Damped Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.05
ar = b*v*abs(v)
a = (-k*dx - m*g - ar)/m


while t < (10):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    ar = b*v*abs(v)
    a = (-k*dx - m*g - ar)/m
    GPE = m*g*x
    EPE = k*dx*dx/2
    KE = m*v*v/2
    TE = GPE + KE + EPE
    t = t + dt
    plt.plot(t, TE, '.', color = 'lime')
    plt.plot(t, GPE, '.', color = 'springgreen')
    plt.plot(t, KE, '.', color = 'aquamarine')
    plt.plot(t, EPE, '.', color = 'turquoise')
    
# re-plotting the last point on each graph to print legend without it looping
plt.plot(t, TE, '.', color = 'lime', label = 'TME')
plt.plot(t, GPE, '.', color = 'springgreen', label = 'GPE')
plt.plot(t, KE, '.', color = 'aquamarine', label = 'KE')
plt.plot(t, EPE, '.', color = 'turquoise', label = 'EPE')
plt.legend()
plt.title("Energy vs. Time")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
```

### Work done by Air Resistance

```python
g = 9.8
v = 0
t = 0
dt = 0.001
ar = b*v*abs(v)
a = (-k*dx - m*g)/m

ds = 0

# making array wih force of air resistance
arlist = []
# making array wih distance traveled
dslist = []
# making array to find the integral of force vs distance (riemann sum, ar * ds)
intlist = []

while t < (0.7):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    ar = b*v*abs(v)
    a = (-k*dx - m*g - ar)/m
    ds = ds + abs(v*dt)
    arlist.append(ar)
    dslist.append(ds)
    intlist.append(abs(ar*abs(v*dt)))
    t = t + dt
    plt.plot(ds, ar, '.', color = 'lightgrey')
plt.xlabel("Distance Traveled (m)")
plt.ylabel("Force of Air Resistance")
plt.fill_between(dslist, arlist, 0, color = 'yellow', label = "work")
plt.legend()
work = sum(intlist)
print("Total work:", round(work,6), "J")
plt.title("Force vs. Distance")
```
