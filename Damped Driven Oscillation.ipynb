```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

### Motion of A Damped Driven Oscillating System

Experimental Constants

```python
# mass of object (kg)
m = 0.029
# spring constant of spring in system (N/m)
k = 3
# initial position of object (m)
x = 0
# initial displacement of spring from unstretched position (m)
dx = 0
# air resistance constant
b = 0.01
# driving force amplitude
F = 0.05
# driving force frequency (Hz)
w = 10
```

Motion of a Vertical Damped Driven Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.01
a = (-k*dx - m*g)/m
ar = b*v*abs(v)
while t < (15):
    v = v + a*dt
    x = x + v*dt 
    dx = x
    ar = b*v*abs(v)
    drive = F*math.sin(w*t)
    a = (-k*dx - m*g - ar + drive)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

Motion of a Horizontal Damped Driven Oscillator

```python
v = 0
t = 0
dt = 0.01
a = (-k*dx)/m
ar = b*v*abs(v)
while t < (15):
    v = v + a*dt
    x = x + v*dt 
    dx = x
    ar = b*v*abs(v)
    drive = F*math.sin(w*t)
    a = (-k*dx - ar + drive)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

### Computationally-determined Resonance Curve for an Oscillating System with up to 5 Different b-values

Experimental Constants

```python
m = 0.0225
k = 4
b1 = 0.08
b2 = 0.04
b3 = 0.02
b4 = 0.01
b5 = 0.005
F = 0.1
```

```python
w = 0

while w < (16):
    A1 = F/(((k-m*w*w)*(k-m*w*w)+(b1*b1*w*w))**(1/2))
    A2 = F/(((k-m*w*w)*(k-m*w*w)+(b2*b2*w*w))**(1/2))
    A3 = F/(((k-m*w*w)*(k-m*w*w)+(b3*b3*w*w))**(1/2))
    A4 = F/(((k-m*w*w)*(k-m*w*w)+(b4*b4*w*w))**(1/2))
    A5 = F/(((k-m*w*w)*(k-m*w*w)+(b5*b5*w*w))**(1/2))
    w = w + 0.01
    plt.plot(w, A1, '.', color = 'darkred')
    plt.plot(w, A2, '.', color = 'firebrick')
    plt.plot(w, A3, '.', color = 'r')
    plt.plot(w, A4, '.', color = 'tomato')
    plt.plot(w, A5, '.', color = 'lightsalmon')
plt.plot(w, A1, '.', color = 'darkred',  label = "b1")
plt.plot(w, A2, '.', color = 'firebrick',  label = "b2")
plt.plot(w, A3, '.', color = 'r',  label = "b3")
plt.plot(w, A4, '.', color = 'tomato',  label = "b4")
plt.plot(w, A5, '.', color = 'lightsalmon',  label = "b5")
plt.legend()

plt.title("Resonance Curve")
plt.xlabel("Driving Force Frequency (Hz)")
plt.ylabel("Amplitude (m)")
```
