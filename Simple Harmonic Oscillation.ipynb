```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

### Motion of A Harmonic Oscillating System

Experimental Constants

```python
# mass of object (kg)
m = 0.029
# spring constant of spring in system (N/m)
k = 3
# initial position of object (m)
x = 0
# unstretched position of spring (m)
xrest = 0
# initial displacement of spring from unstretched position (m)
dx = 0
```

Motion of a Vertical Harmonic Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.01
a = (-k*dx - m*g)/m

while t < (2):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    a = (-k*dx - m*g)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

Motion of a Horizontal Harmonic Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.01
a = (-k*dx)/m

while t < (2):
    v = v + a*dt
    x = x + v*dt 
    dx = x-xrest
    a = (-k*dx)/m
    t = t + dt
    plt.plot(t, x, '.', color = 'black')
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
```

### Energy of a Harmonic Oscillating System

Experimental Constants

```python
# mass of object (kg)
m = 0.029
# spring constant of spring in system (N/m)
k = 3
# initial position of object (m)
x = 0
# unstretched position of spring (m)
xrest = 0
# initial displacement of spring from unstretched position (m)
dx = 0
```

Energy of a Vertical Harmonic Oscillator

```python
g = 9.8
v = 0
t = 0
dt = 0.05
a = (-k*dx - m*g)/m

while t < (10):
    v = v + a*dt
    x = x + v*dt
    dx = x-xrest
    a = (-k*dx - m*g)/m
    GPE = m*g*x
    EPE = k*dx*dx/2
    KE = m*v*v/2
    TE = GPE + KE + EPE
    t = t + dt
    plt.plot(t, TE, '.--', color = 'deeppink')
    plt.plot(t, GPE, '.--', color = 'blue')
    plt.plot(t, KE, '.--', color = 'blueviolet')
    plt.plot(t, EPE, '.--', color = 'magenta')

# re-plotting the last point on each graph to print legend without it looping
plt.plot(t, TE, '.', color = 'deeppink', label = 'TME')
plt.plot(t, GPE, '.', color = 'blue', label = 'GPE')
plt.plot(t, KE, '.', color = 'blueviolet', label = 'KE')
plt.plot(t, EPE, '.', color = 'magenta', label = 'EPE')
plt.legend()
plt.title("Energy vs. Time")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
```
